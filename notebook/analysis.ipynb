```python
# importing package
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import json
import carball
from carball.json_parser.game import Game
from carball.analysis.analysis_manager import AnalysisManager
```

```python

# List of replay file paths
replay_files = ['../demoFiles/NRGvsG2/5698B1B6487CF73EE305F08B6D504D71.replay',
                '../demoFiles/NRGvsG2/2A66A21F4AD82750087A80BC7DC3D951.replay',
                '../demoFiles/NRGvsG2/9A83EF694E78754A06F2569DC1501B53.replay',
                '../demoFiles/NRGvsG2/9BEAED8E48E1B4C0054F8499D649C390.replay',
                '../demoFiles/NRGvsG2/341AAB7F422FEA52E9FDF48A6E535301.replay',
                '../demoFiles/NRGvsG2/715312F94C58EC5E3F02AEBF4C22D451.replay',
                '../demoFiles/NRGvsG2/D83190474AB0043E7595FDB3E1EC12E0.replay']

# List to store the JSON objects
json_objects = []


# Iterate over the replay files
for i, replay_file in enumerate(replay_files):
    
    # Decompile the replay file and retrieve the JSON data
    _json = carball.decompile_replay(replay_file)

    # Initialize a Game object and load the JSON data
    game = Game()
    game.initialize(loaded_json=_json)

    # Create an AnalysisManager object for the game
    analysis_manager = AnalysisManager(game)
    analysis_manager.create_analysis()

    # Get the protobuf data
    proto_object = analysis_manager.get_protobuf_data()

    # return the proto object as a json object
    json_object = analysis_manager.get_json_data()
    
    # Append the json_object to the list
    json_objects.append(json_object)  
    
```

```python
# Extracting the metadata for each team, which includes player IDs, team names, and team colors.

index = 0
dataset_teamMetaData = []  # List to store the metadata for each team
teams_Metadata = []  # List to store individual team metadata

# Iterating through replay files
for i in enumerate(replay_files):
    json_object = json_objects[index]  # Accessing the JSON object for the current replay file
    index = index + 1
    
    # Extracting team metadata using json_normalize
    data = pd.json_normalize(json_object['teams'], record_path=['playerIds'], meta=['name', 'score', 'isOrange'])
    dataset_teamMetaData.append(data)  # Adding the team metadata to the dataset
    
    # Creating a DataFrame for team metadata and dropping the 'score' column
    Team_meta = pd.DataFrame(data)
    Team_meta = Team_meta.drop(columns=['score'])
    teams_Metadata.append(Team_meta)  # Adding the team metadata DataFrame to the list
```

```python
# Extracting Goals data of each team, which includes frame number and Player ID.

goals_Data = []  # List to store the goals data for each team

# Iterating through the replay files
for i in range(len(replay_files)):
    json_object = json_objects[i]  # Accessing the JSON object for the current replay file
    
    # Extracting goals data using json_normalize
    data = pd.json_normalize(json_object['gameMetadata'], record_path=['goals'])
    
    GMD_goals = pd.DataFrame(data)
    GMD_goals.insert(0, 'total_goals', range(1, 1 + len(GMD_goals)))  # Adding a column for total goals
    GMD_goals = pd.merge(GMD_goals, teams_Metadata[i], left_on='playerId.id', right_on='id', how='left')  # Merging with team metadata
    GMD_goals = GMD_goals.drop(GMD_goals.columns[2], axis=1)  # Dropping the duplicate 'id' column
    
    goals_Data.append(GMD_goals)  # Adding the goals data DataFrame to the list
```

```python
# Extracting each Hit data and merging it with Team.

hits = []  # List to store the hit data for each team
hits_and_teams = []  # List to store the hit data merged with team metadata

# Iterating through the replay files
for i in range(len(replay_files)):
    json_object = json_objects[i]  # Accessing the JSON object for the current replay file
    
    # Extracting hit data using json_normalize
    data = pd.json_normalize(json_object['gameStats'], record_path=['hits'])
    
    GS = pd.DataFrame(data)
    GS_hits = GS.drop(GS.columns[[1, 2, 3, 8, 9, 10]], axis=1)  # Selecting relevant columns and dropping unnecessary ones
    GS_hits.rename(columns={'playerId.id': 'id'}, inplace=True)  # Renaming the 'playerId.id' column to 'id'
    hits.append(GS_hits)  # Adding the hit data DataFrame to the list
    
    # Merging hit data with team metadata
    Hits_merged_with_Teams = pd.merge(GS_hits, teams_Metadata[i], on='id', how='left')
    
    # Adding a 'Team' column to indicate the team (0 for blue, 1 for orange)
    Hits_merged_with_Teams['Team'] = np.where(Hits_merged_with_Teams['isOrange'] != True, 0, 1)
    
    hits_and_teams.append(Hits_merged_with_Teams)  # Adding the merged DataFrame to the list
```

```python
# Extracting Demos data of each team, which includes frame number, demos count, attacker id, victim id, team name.

demos_Data = []

for i in range(len(replay_files)):
    
    json_object = json_objects[i]
    
    data = pd.json_normalize(json_object['gameMetadata'], record_path= ['demos'])
    GMD_demos = pd.DataFrame(data)
    GMD_demos.insert(0, 'total_demos', range(1, 1 + len(GMD_demos)))
    GMD_demos= pd.merge(GMD_demos, teams_Metadata[i], left_on='victimId.id', right_on='id', how='left')
    
    demos_Data.append(GMD_demos)
    
#print(demos_Data[6].head(4))    
```

Q2. What is the correlation between demolitions in the last 20% of a match and its outcome, and how does this influence the final result of the game?
</h2>

```python

# Calculating the starting and ending frames in all games.

frames_start  = []
frames_end = []

for i in range(len(replay_files)):
    
    json_object = json_objects[i]
    data = pd.json_normalize(json_object['gameMetadata'])
    gameMetadata = pd.DataFrame(data)
    
    frames = gameMetadata['frames'][0]

    f2 = (20*frames)/100   #final 20 precent of the game  
    st = frames-f2         # staring frame 
    end = frames           #ending frame
    frames_start.append(st)
    frames_end.append(end)


```

```python
#Calculating the number of goals and demoes scored in the final 20 percent of the match

Tb = "NRG ESPORTS"
To = "G2 ESPORTS"
winner_team = []
blue_Goals_F20 = []
orange_Goals_F20 = []
blue_Demos_F20 = []
orange_Demos_F20 = []


for x in range(len(replay_files)):
    
    #calculating goals and winning team name
    countb=0
    counto =0
    for i in range(len(goals_Data[x])):
        
            TB_goals_f20 = 0
            TO_goals_f20 = 0
            TB_demos_f20 = 0
            TO_demos_f20 = 0
            
            if goals_Data[x].loc[i, 'frameNumber'] >= frames_start[x]:
                if goals_Data[x].loc[i,'name'] == Tb:
                    TB_goals_f20 = TB_goals_f20+1
                elif goals_Data[x].loc[i,'name'] == To:
                    TO_goals_f20 = TO_goals_f20+1
   
    countb = goals_Data[x]['name'].value_counts().get(Tb, 0)
    counto = goals_Data[x]['name'].value_counts().get(To, 0)

    if(countb > counto):
            winner = "Team Blue Won"
            winner_team.append(winner)
   
    else:
            winner = "Team Orange Won"
            winner_team.append(winner)
    
    blue_Goals_F20.append(TB_goals_f20)
    orange_Goals_F20.append(TO_goals_f20)
            
        
    #calculating demos
    
    for i in range(len(demos_Data[x])):
        if demos_Data[x].loc[i, 'frameNumber'] >= frames_start[x]:
            if demos_Data[x].loc[i,'name'] == Tb:
                TO_demos_f20 = TO_demos_f20+1                    #Team orange demoed team blue // team blue demos
            elif demos_Data[x].loc[i,'name'] == To:
                TB_demos_f20 = TB_demos_f20+1

    blue_Demos_F20.append(TB_demos_f20)
    orange_Demos_F20.append(TO_demos_f20)
```
